To get players' draft positions, we need the CommonPlayerInfo endpoint. This endpoint times out ~ 5% of the time, so we need to write a retry wrapper. The code is borrowed from the [NBA_api documentation](https://github.com/swar/nba_api/blob/master/docs/examples/Home%20Team%20Win-Loss%20Modeling/Home%20Team%20Win-Loss%20Data%20Prep.ipynb)

Even with the retry wrapper, the API fails often. Instead of repeatedly trying the same queries, we will iteratively plug the gaps in the data until 

```python
# Retry Wrapper 
def retry(func, retries=2):
    def retry_wrapper(*args, **kwargs):
        attempts = 0
        while attempts < retries:
            try:
                return func(*args, **kwargs)
            except requests.exceptions.RequestException as e:
                print(e)
                time.sleep(30)
                attempts += 1

    return retry_wrapper

@retry
def process_player(player):
    print(player['full_name'])
    res = commonplayerinfo.CommonPlayerInfo(player_id=player['id']).get_data_frames()[0]
    res.loc[:,'PLAYER_NAME'] = player['full_name']
    return res   

#Rick Adelman

res_df_1 = pd.concat([process_player(player) for player in nba_players])
```

CommonPlayerInfo does have player position, but it would take some data engineering to get the right position for the right season. Our kaggle dataset has done the data engineering for us, so we can more easily get position from that

```python
def get_pos_list(pos):
    positions = pos.split('-')
    position_set = set()
    
    if ('F' in positions) or ('SF' in positions):
        position_set.add('SF')  
        
    if ('F' in positions) or ('PF' in positions):
        position_set.add('PF') 
        
    if ('G' in positions) or ('SG' in positions):
        position_set.add('SG')            
        
    if ('G' in positions) or ('PG' in positions):
        position_set.add('PG') 
        
    if ('C' in positions):
        position_set.add('C')    
    return position_set

player_data = pd.read_csv('../data/raw/player_data.csv') #downloaded from kaggle

player_data['player'] = [cleanup_name_str(x) for x in player_data['player']]

player_data.loc[:,'pos'] = [get_pos_list(pos) for pos in player_data['pos']] 
positions = player_data.groupby(['player','season'])['pos'].aggregate(lambda x: reduce(set.union, x))
positions.to_csv('../data/positions.csv')

player_id_reference = pd.DataFrame(players.get_players())
player_id_reference.rename(columns = {'full_name' : 'player'}).to_csv('../data/player_id_reference.csv', index = False)
```

```python
start = datetime.now()
seasons = ['2000-01','2001-02','2002-03']
for season in seasons:
    gamefinder = leaguegamefinder.LeagueGameFinder(season_nullable=season, season_type_nullable = 'Regular Season')
    games_and_dates = gamefinder.get_data_frames()[0].groupby(['GAME_ID','GAME_DATE']).count().index
    box_scores_all_games = pd.concat([box_score_helper(g, date) for g, date in games_and_dates])
    box_scores_all_games.to_csv('../data/stat_data/' + season + '.csv', index = False)
print(datetime.now() - start)
```
